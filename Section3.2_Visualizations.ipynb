```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel("Data_Visualization.xlsx")
DF = df.set_index('Time')
DF.rename(columns = {'GDP ':'GDP'}, inplace = True)
DF.columns[:-1]
DF
```

```python
for i in DF.columns[:-1]:
    plt.figure(figsize=(12,5))
    plt.xlabel(f'Trade Openness vs. {i}')

    ax1 = DF['Trade Openness'].plot(color='blue', grid=True, label='Trade Openness')
    ax2 = DF[i].plot(color='red', grid=True, secondary_y=True, label=f'{i}')

    ax1.legend(loc=2)
    ax2.legend(loc=1)

    plt.savefig(f"OpennessVS{i}.jpg")
    plt.show()
```
